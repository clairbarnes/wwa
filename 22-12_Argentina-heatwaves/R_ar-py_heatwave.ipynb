### **LR tests of number of parameters needed**
- in this study, including a non-zero shape parameter doesn't appreciably improve the fit
- shifting the distribution brings a significant improvement, but scaling it does not.

```r
fit_sgumbel <- fevd(x=tmax, data=df, type="Gumbel")
fit_nsgumbel <- fevd(x=tmax, data=df, location.fun=~gmst, type="Gumbel")
fit_ssgumbel <- fevd(x=tmax, data=df, location.fun=~gmst, scale.fun=~gmst, type="Gumbel")

fit_sgev <- fevd(x=tmax, data=df, type="GEV")
fit_nsgev <- fevd(x=tmax, data=df, location.fun=~gmst, type="GEV")
fit_ssgev <- fevd(x=tmax, data=df, location.fun=~gmst, scale.fun=~gmst, type="GEV")
```

```r
data.frame(list("Gumbel shift" = unname(lr.test(fit_sgumbel, fit_nsgumbel)$p.value),
                "Gumbel shift+scale" = unname(lr.test(fit_nsgumbel, fit_ssgumbel)$p.value),
                "GEV shift" = unname(lr.test(fit_sgev, fit_nsgev)$p.value),
                "GEV or Gumbel shift" = unname(lr.test(fit_nsgumbel, fit_nsgev)$p.value),
                "Gumbel shift+scale" = unname(lr.test(fit_nsgumbel, fit_ssgumbel)$p.value)))
```

```r
df["gmst2"] <- df$gmst^2
fit_nsgumbel_quadratic <- fevd(x=tmax, data=df, location.fun=~gmst2, type="Gumbel")
# can't use LR test because same number of parameters in each model
# AIC  is slightly lower for quadratic model
```

### **Maps of fitted parameters**


```r
era5 = read.csv("era5-flattened.csv", skip = 3, header = F)
colnames(era5)[1] <- "year"
era5 <- merge(read.csv("../90_data/gmst.csv"), era5)
```

#### **Fit the models**

```r
system.time({
    mle_gev = lapply(colnames(era5)[-(1:2)], function(cnm) {
        if(is.na(era5[1,cnm])) {
            return(NULL)
        } else {
            return(fevd(x=tmax, data=setNames(era5[,c("gmst",cnm)], c("gmst", "tmax")), location.fun=~gmst, type="GEV", optim.args = list(method = "Nelder-Mead")))
        }
    })
})
```

#### **Fitted parameters, exceedances etc**

```r
ml_pars <- data.frame(t(do.call("rbind", lapply(1:length(mle_gev), function(i) {
    if(is.null(mle_gev[[i]])) {
        return(setNames(rep(NA,4), c("mu0", "mu1", "sigma0", "shape")))
    } else {
        model_pars(mle_gev[[i]])[-4]
    }
}))))
                     
ml_rv <- data.frame(t(do.call("rbind", lapply(1:length(mle_gev), function(i) {
    if(is.null(mle_gev[[i]])) {
        return(c("ep" = NA, "rp" = NA, "rl10" = NA, "rl20" = NA, "pr" = NA, "delta_I" = NA))
    } else {
        mdl <- mle_gev[[i]]
        qc = event_qcov(mdl, event_gmst)

        ev = mdl$x[73]
        ep = 1-pextRemes(mdl, q = ev, qcov = qc)
        rp = 1/(ep)
        rl10 = return_level(mdl, 10, event_gmst)
        rl20 = return_level(mdl, 20, event_gmst)
        pr = prob_ratio(mdl, ev, event_gmst, event_gmst - 1.2)
        di = delta_I(mdl, rp, event_gmst, event_gmst - 1.2)
        
        return(c("ep" = ep, "rp" = rp, "rl10" = rl10, "rl20" = rl20, "pr" = pr, "delta_I" = di))
    }
}))))

ml_res = rbind(ml_pars, ml_rv)
write.csv(ml_res, "era5-flattened_ml-ests.csv")
```

#### **Get parameter estimates, exceedances etc**

```r
# load the fitted models
mh_gev = unlist(lapply(1:36, function(a) {
    load(paste0("fitted_models/mh_gev_",formatC(a, width = 2, flag = "0"), ".Rdata"))
    return(mh_gev)
}), recursive = F)
```

```r
# parameter estimates
mh_pars <- data.frame(t(do.call("rbind", lapply(1:length(mh_gev), function(i) {
    if(is.null(mh_gev[[i]])) {
        return(setNames(rep(NA,4), c("mu0", "mu1", "sigma0", "shape")))
    } else {
        model_pars(mh_gev[[i]])[-4]
    }
}))))

# event value & return period                    
mh_rv <- data.frame(t(do.call("rbind", lapply(1:length(mh_gev), function(i) {
    if(is.null(mh_gev[[i]])) {
        return(c("ep" = NA, "rp" = NA, "rl10" = NA, "rl20" = NA, "pr" = NA, "delta_I" = NA))
    } else {
        mdl <- mh_gev[[i]]
        
        qc = event_qcov(mdl, event_gmst)
        
        ev = mdl$x[73]
        ep = 1-pextRemes(mdl, q = ev, qcov = qc)
        rp = 1/(ep)
        rl10 = return_level(mdl, 10, event_gmst)
        rl20 = return_level(mdl, 20, event_gmst)
        pr = prob_ratio(mdl, ev, event_gmst, event_gmst - 1.2)
        
        if(is.finite(rp)) {
            di = delta_I(mdl, rp, event_gmst, event_gmst - 1.2)
        } else {
            di = NA
        }
        
        return(c("ep" = ep, "rp" = rp, "rl10" = rl10, "rl20" = rl20, "pr" = pr, "delta_I" = di))
    }
}))))

# will use profile likelihood test to check whether nonzero shape parameter is actually meaningful

# output results to dataframe
mh_res = rbind(mh_pars, mh_rv)
write.csv(mh_res, "era5-flattened_bayesian-ests.csv")
```
